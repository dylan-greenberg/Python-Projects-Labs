```python
df['public_metrics.impression_count'] = pd.to_numeric(df['public_metrics.impression_count'])
df['public_metrics.like_count'] = pd.to_numeric(df['public_metrics.like_count'])
df['public_metrics.reply_count'] = pd.to_numeric(df['public_metrics.reply_count'])
df['public_metrics.retweet_count'] = pd.to_numeric(df['public_metrics.retweet_count'])
df.drop(['reply_settings', 'attachments.poll_ids', 'attachments.media_source_tweet_id', 'entities.cashtags', 'withheld.country_codes'], axis=1, inplace=True)
df.info()
```

## Summarize Data

‚û°Ô∏è Assignment Tasks
- Create at least two charts to summarize and visualize the data
- Examples include: Number of posts per device; top 10 posts with highest engagement

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by like count in descending order and get the top 10
top_10_tweets = df.sort_values('public_metrics.like_count', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_tweets['username'], top_10_tweets['public_metrics.like_count'])
plt.xlabel('Author Username')
plt.ylabel('Number of Likes')
plt.title('Top 10 Tweets with Most Likes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort by retweet count in descending order
df_sorted = df.sort_values(by='public_metrics.retweet_count', ascending=False)

# Select the top N tweets (e.g., top 10)
top_tweets = df_sorted.head(10)

# Plot the number of retweets (or simply use the index for bar heights)
plt.figure(figsize=(12, 10))
bars = plt.barh(top_tweets['username'].astype(str), top_tweets['public_metrics.retweet_count'], color='skyblue')

# Annotate with tweet text
for bar, text in zip(bars, top_tweets['text']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, text,
             va='center', ha='left', fontsize=8, color='black', wrap=True)

plt.title('Top 10 Tweets')
plt.xlabel('Number of Retweets')
plt.ylabel('Username')
plt.tight_layout()
plt.show()
```

```python


for index, row in top_tweets.iterrows():
    print(f"Username: {row['username']}")
    print(f"Retweets: {row['public_metrics.retweet_count']}")
    print(f"Text: {row['text']}")
    print("\n" + "-"*50 + "\n")
```

```python
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Select the top 500 tweets
df_sorted = df.sort_values(by='public_metrics.like_count', ascending=False)
top_500_tweets = df_sorted.head(500)

# Combine all tweet texts into a single string
text = ' '.join(top_500_tweets['text'].dropna())

# Remove specific words (e.g., 'https')
text = re.sub(r'https?\S+', '', text)  # Remove URLs
text = re.sub(r'\bhttps\b', '', text)  # Remove the word 'https'

# Remove single-letter words
text = ' '.join(word for word in text.split() if len(word) > 1)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Top 500 Most Liked Tweets (Filtered)')
plt.show()
```
